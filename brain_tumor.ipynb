Checking out a sample image from the dataset.

```python
import cv2
import matplotlib.pyplot as plt
import os

# Load an MRI image
image_path = 'brain_tumor_dataset/no/1 no.jpeg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Display using Matplotlib
plt.imshow(image, cmap='gray')
plt.title(" Sample MRI Image")
plt.axis('off')
plt.show()
```

Distribution of "yes" and "no" labels.

```python
yes_dir = 'brain_tumor_dataset/yes/'
no_dir = 'brain_tumor_dataset/no/'

num_yes = len(os.listdir(yes_dir))
num_no = len(os.listdir(no_dir))

print(f'Number of "yes" images: {num_yes}')
print(f'Number of "no" images: {num_no}')

# Visualize distribution using Matplotlib
labels = ['Yes', 'No']
counts = [num_yes, num_no]

plt.bar(labels, counts, color=['red', 'blue'])
plt.title("Distribution of MRI Images")
plt.ylabel('Number of Images')
plt.show()
```

Normalizing pixel values to the range [0,1].

```python
source_directory = 'brain_tumor_dataset/no/'
target_directory = 'no_normalized/'

# Get a list of all images in the source directory
image_filenames = [f for f in os.listdir(source_directory) if os.path.isfile(os.path.join(source_directory, f))]

for image_filename in image_filenames:
    image_path = os.path.join(source_directory, image_filename)
    
    # Read and normalize the image
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    normalized = image / 255.0
    
    # Convert the normalized image back to the range 0-255
    output_image = (normalized * 255).astype('uint8')
    
    # Save the image to the target directory
    output_path = os.path.join(target_directory, image_filename)
    cv2.imwrite(output_path, output_image)
```

```python
source_directory = 'brain_tumor_dataset/yes/'
target_directory = 'yes_normalized/'

# Get a list of all images in the source directory
image_filenames = [f for f in os.listdir(source_directory) if os.path.isfile(os.path.join(source_directory, f))]

for image_filename in image_filenames:
    image_path = os.path.join(source_directory, image_filename)
    
    # Read and normalize the image
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    normalized = image / 255.0
    
    # Convert the normalized image back to the range 0-255
    output_image = (normalized * 255).astype('uint8')
    
    # Save the image to the target directory
    output_path = os.path.join(target_directory, image_filename)
    cv2.imwrite(output_path, output_image)
```

Resizing images to 224 x 224.

```python
source_directory = 'no_normalized'
target_size = (224, 224) 

# Loop over each file in the directory
for image_filename in os.listdir(source_directory):
    image_path = os.path.join(source_directory, image_filename)
    
    # Check if the path points to a file (to exclude directories, if any)
    if os.path.isfile(image_path):
        
        # Read the image
        image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # or cv2.IMREAD_COLOR if they are RGB images
        
        # Resize the image
        resized_image = cv2.resize(image, target_size)
        
        # Save the resized image back to the same location, overwriting the original
        cv2.imwrite(image_path, resized_image)
```
